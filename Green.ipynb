```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('greendestination (1).csv')

df.head()
```

```python
#Attrition rate

attrition_count = df[df['Attrition'] == 'Yes'].shape[0]
total_employees = len(df)
 
attrition_rate = (attrition_count / total_employees) * 100

print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
#Age

my_colors = {"Yes": "pink", "No": "purple"}---
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(data=df, x='Age', hue='Attrition', multiple='stack', palette=my_colors)
plt.title('Age (Red = Left)')
```

```python
#Year at company

plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 2)
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', multiple='stack', palette=my_colors)
plt.title('Years at Company')
```

```python
#Inco

plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 3)
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, hue='Attrition', palette=my_colors, legend=False)
plt.title('Monthly Income')

plt.tight_layout()
plt.show()
```
